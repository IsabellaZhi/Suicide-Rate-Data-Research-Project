#### 6.2.2 Hypothesis Testing with Bootstrap

Now we will be testing our model under the null hypothesis that `GDP per capita` and `suicide rate by age` are not correlated to each other. To do this, we will perform a bootstrap resampling of our data and find a 95% confidence interval for the slopes of our resampled data. 

```python
np.random.seed(2950)
n_bootstrap = 500
bootstrap_slopes10 = np.zeros(n_bootstrap)
bootstrap_slopes20 = np.zeros(n_bootstrap)
bootstrap_slopes30 = np.zeros(n_bootstrap)
bootstrap_slopes40 = np.zeros(n_bootstrap)
bootstrap_slopes50 = np.zeros(n_bootstrap)
bootstrap_slopes60 = np.zeros(n_bootstrap)
bootstrap_slopes70 = np.zeros(n_bootstrap)
bootstrap_slopes80 = np.zeros(n_bootstrap)

for i in range(n_bootstrap):
    resampled_data = crude_suicide_clean_sorted.sample(n=12, replace=True)
    resampled_model10 = LinearRegression().fit(resampled_data[["GDP"]], resampled_data['10to19'])
    resampled_model20 = LinearRegression().fit(resampled_data[["GDP"]], resampled_data['20to29'])
    resampled_model30 = LinearRegression().fit(resampled_data[["GDP"]], resampled_data['30to39'])
    resampled_model40 = LinearRegression().fit(resampled_data[["GDP"]], resampled_data['40to49'])
    resampled_model50 = LinearRegression().fit(resampled_data[["GDP"]], resampled_data['50to59'])
    resampled_model60 = LinearRegression().fit(resampled_data[["GDP"]], resampled_data['60to69'])
    resampled_model70 = LinearRegression().fit(resampled_data[["GDP"]], resampled_data['70to79'])
    resampled_model80 = LinearRegression().fit(resampled_data[["GDP"]], resampled_data['80_above'])
   
    bootstrap_slopes10[i] = resampled_model10.coef_[0]
    bootstrap_slopes20[i] = resampled_model20.coef_[0]
    bootstrap_slopes30[i] = resampled_model30.coef_[0]
    bootstrap_slopes40[i] = resampled_model40.coef_[0]
    bootstrap_slopes50[i] = resampled_model50.coef_[0]
    bootstrap_slopes60[i] = resampled_model60.coef_[0]
    bootstrap_slopes70[i] = resampled_model70.coef_[0]
    bootstrap_slopes80[i] = resampled_model80.coef_[0]
   
ci10 = np.quantile(bootstrap_slopes10, q=[0.025, 0.975])
ci20 = np.quantile(bootstrap_slopes20, q=[0.025, 0.975])
ci30 = np.quantile(bootstrap_slopes30, q=[0.025, 0.975])
ci40 = np.quantile(bootstrap_slopes40, q=[0.025, 0.975])
ci50 = np.quantile(bootstrap_slopes50, q=[0.025, 0.975])
ci60 = np.quantile(bootstrap_slopes60, q=[0.025, 0.975])
ci70 = np.quantile(bootstrap_slopes70, q=[0.025, 0.975])
ci80 = np.quantile(bootstrap_slopes80, q=[0.025, 0.975])

'10to19', '20to29', '30to39', '40to49', '50to59','60to69','70to79','80_above'

print(f'95% confidence interval of the regression slope for age group 10 to 19: {ci10[0]:.8f} to {ci10[1]:.8f}')
print(f'95% confidence interval of the regression slope for age group 20 to 29: {ci20[0]:.8f} to {ci20[1]:.8f}')
print(f'95% confidence interval of the regression slope for age group 30 to 39: {ci30[0]:.8f} to {ci30[1]:.8f}')
print(f'95% confidence interval of the regression slope for age group 40 to 49: {ci40[0]:.8f} to {ci40[1]:.8f}')
print(f'95% confidence interval of the regression slope for age group 50 to 59: {ci50[0]:.8f} to {ci50[1]:.8f}')
print(f'95% confidence interval of the regression slope for age group 60 to 69: {ci60[0]:.8f} to {ci60[1]:.8f}')
print(f'95% confidence interval of the regression slope for age group 70 to 79: {ci70[0]:.8f} to {ci70[1]:.8f}')
print(f'95% confidence interval of the regression slope for age group 80 and above: {ci80[0]:.8f} to {ci80[1]:.8f}')
```

### 6.3 The relationship between `suicide rate` and `age`

#### 6.3.1 Regression Model 2 `Linear Model`

```python
crude_suicide_both_sexes = crude_suicide[crude_suicide['Sex'] == ' Both sexes']
```

```python
#removing the country and sex columns in the data frame
crude_suicide_both_sexes=crude_suicide_both_sexes[[' 80_above',' 70to79', ' 60to69 ', ' 50to59 ', ' 40to49', ' 30to39', ' 20to29', ' 10to19']]

#stacking the suicide rate values so it is all in one column
crbs_stack_vals=crude_suicide_both_sexes[[' 80_above',' 70to79', ' 60to69 ', ' 50to59 ', ' 40to49', ' 30to39', ' 20to29', ' 10to19']].stack()
```

```python
#making the series into a data frame with one of the indices as a column 
crbs=pd.DataFrame()
crbs['suicide rate']=crbs_stack_vals
crbs['ages']=crbs.index.get_level_values(1)
crbs
```

```python
plt.scatter(crbs['ages'], crbs['suicide rate'], alpha = 0.6)
plt.xlabel('age ranges')
plt.ylabel('suicide rate (per 100,000 individuals)')
plt.title('Suicide Rate by Age Ranges');
```

```python
#Making the age values into dummy variables to be able to perform linear regression
age_map={" 80_above": 8, 
         " 70to79": 7, 
         " 60to69 ": 6, 
         " 50to59 ": 5, 
         " 40to49": 4 , 
         " 30to39":3, 
         " 20to29":2, 
         " 10to19":1}

crbs['ages_mapped']=crbs['ages'].map(age_map)
crbs
```

```python
#fitting a linear regression model
from sklearn.linear_model import LinearRegression

age_model=LinearRegression().fit(crbs[['ages_mapped']], crbs['suicide rate'])
sns.regplot(crbs[['ages_mapped']], y=crbs['suicide rate'], scatter_kws={'alpha': 0.05})
plt.title('Linear Regression of Suicide Rate by Mapped Age Ranges')
plt.ylabel('suicide rate (per 100,000 individuals)');
```

```python
slope= age_model.coef_[0]
r2= age_model.score(crbs[['ages_mapped']], crbs['suicide rate'])

print('slope of the model is:', round(slope, 2))
print('r^2:', round(r2, 2))
```

**The linear model shows that there is a positive correlation with suicide rate and age. As people become older, suicide rate tends to increase as well. This is shown in the slope of the model, which is 4.06. The r^2 value of the model is very small, which means the linear regression model does not fit the data very well. However, this is only the case because the age data is categorical, so when plotting the data, it shows data points in columns. Since the data points are not continuous, the linear model shows a low r^2 value. However, this does not necessarily mean that the linear model is insignificant as it shows a positive correlation between suicide rate and age.** 

#### 6.3.2 Hypothesis Testing with Bootstrap for `Linear Model`

Now we will be testing our model under the null hypothesis that age and suicide rate are not correlated to each other. To do this, we will perform a bootstrap resampling of our data and find a 95% confidence interval for the slopes of our resampled data. 

```python
np.random.seed(2950)
n_bootstrap = 500
bootstrap_slopes = np.zeros(n_bootstrap)

for i in range(n_bootstrap):
    resampled_data = crbs.sample(n=len(crbs), replace=True)
    resampled_model = LinearRegression().fit(resampled_data[['ages_mapped']], resampled_data['suicide rate'])
   
    bootstrap_slopes[i] = resampled_model.coef_[0]


# 95% confidence interval:
print('confidence interval: ', np.round(np.quantile(bootstrap_slopes, q=[0.025, 0.975]),2))
```

This shows that with 95% confidence, the slope of the Linear Regression will not be equal to zero, which means that with 95% confidence, we can reject the null hypothesis that there will be no correlation between age and suicide rate. This validates the conclusion that there is a relationship between age and suicide rate. 

#### 6.3.3 Regression Model 3 `Exponential Model`

```python
#making the suicide rate values of 0 into a value of 0.001 to represent the countries with no suicide rates
crbs['suicide rate']=crbs['suicide rate'].replace(0,0.001)

#Fitting an exponentil model
log_suicide=np.log(crbs['suicide rate'])
model_expo=LinearRegression().fit(crbs[['ages_mapped']], log_suicide)
```

```python
slope_expo=model_expo.coef_[0]
r2_expo=model_expo.score(crbs[['ages_mapped']], log_suicide)
c=model_expo.intercept_

print('slope of the exponential model is:', round(slope_expo, 2))
print('r^2:', round(r2_expo, 2))
```

```python
crbs=crbs.sort_values(by='ages_mapped')

y_pred=np.exp(c) * np.exp(slope_expo * crbs['ages_mapped'])

plt.scatter(crbs['ages'], crbs['suicide rate'], alpha=0.3, label='observed data') #observed data
plt.plot(crbs['ages'], y_pred, c='orange', label='predicted data') # predicted data
plt.xlabel('age ranges')
plt.ylabel('suicide rate (per 100,000 individuals)')
plt.title('Exponenetial Model of Suicide Rate by Age Ranges')
plt.legend();
```

**We can see that the exponential model is a worst fit to the data than the linear model is. We had originally predicted that the exponential model would be a better fit, but this has not been the case. The exponential model shows a lower r^2 value than the linear model and has a lower coefficient of correlation. In general, the relationship between age and suicide rate is more linear than it is exponential.**

#### 6.3.4 Hypothesis Testing with Bootstrap for `Exponential Model`

We will now perform bootstrap resampling for the exponential model to test the validity of our data. Our null hypothesis, once again, is that there is no relationship between age range and suicide rate for the exponential model.


```python
np.random.seed(2950)
n_bootstrap = 500
bootstrap_slopes = np.zeros(n_bootstrap)

for i in range(n_bootstrap):
    resampled_data = crbs.sample(n=len(crbs), replace=True)
    resampled_model = LinearRegression().fit(resampled_data[['ages_mapped']], np.log(resampled_data['suicide rate']))
   
    bootstrap_slopes[i] = resampled_model.coef_[0]


# 95% confidence interval:
print('confidence interval: ', np.round(np.quantile(bootstrap_slopes, q=[0.025, 0.975]),2))
```

#### Although the 95% confidence interval shows very small values for the resampled slopes, with 95% confidence, the slope will never equal zero, which means that there is stasticial significance in our exponential model and we can reject the null hypothesis. However, since the slopes are much smaller, this supports our conclusion that the linear model is better than the exponential model for our data, showing a higher correlation between the variables than the exponential model with 95% confidence.
